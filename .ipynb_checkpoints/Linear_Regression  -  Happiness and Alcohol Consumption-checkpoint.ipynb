# Happiness and Alcohol Consumption

- Country;
- Region;
- Hemisphere;
- HappinessScore: Nível de Felicidade;
- HDI (Human Development Index by United Nations Development Programme): Indice de Desenvolvimento Humano (IDH);
- GDP_PerCapita (Gross Domestic Product PerCapita): Produto Interno Bruto por cabeça (PIB);
- Beer_PerCapita (Liters (per capita) of beer consumption): Consumo de Cerveja por cabeça em Litros;
- Spirit_PerCapita (Consumption of spirits drink ( per capita )): Consumo de bebidas espirituosas por cabeça;
- Wine_PerCapita (Wine Consumption): Consumo de Vinho por cabeça.

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

%matplotlib inline
```

Leitura do Dataset:

```python
happy = pd.read_csv("HappinessAlcoholConsumption.csv")
happy.head()
```

Como há algumas variáveis, podemos trabalhar com algumas, como o Consumo de Cerveja e Vinho PerCapita (Beer_PerCapita e Wine_PerCapita), a fim de buscar a relação do consumo com o Score de Felicidade.

```python
# DESCRIÇÃO DO DATASET:
happy.describe()
```

- Verificando algumas relações:

```python
# HAPPINESS SCORE E CONSUMO DE CERVEJA POR CABEÇAA
sns.jointplot(x='Beer_PerCapita', y='HappinessScore', data=happy);
```

```python
# HAPPINESS SCORE E CONSUMO DE VINHO POR CABEÇA
sns.jointplot(x='Wine_PerCapita', y='HappinessScore', data=happy);
```

```python
sns.pairplot(happy);
```

Algo interessante de se observar na plotagem, é a relação linear entre HDI (Indice de Desenvolvimento Humano) e Happiness Score. Ou seja, quanto maior o HDI do País maior é o Score de Felicidade. 

```python
happy.columns
```

['Country', 'Region', 'Hemisphere', 'HappinessScore', 'HDI',
       'GDP_PerCapita', 'Beer_PerCapita', 'Spirit_PerCapita',
       'Wine_PerCapita']

```python
# TRABALHAREMOS  APENAS COM O HDI, Beer_PerCapita e Wine_PerCapita
x = happy[['HDI', 'Beer_PerCapita', 'Wine_PerCapita']]
y = happy['HappinessScore']
```

- Dividindo o dataset em dados de treinamento e dados de teste com train_test_split do scikitlearn

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 110)
```

```python
model = LinearRegression()
```

- TREINANDO O MODELO E ENCONTRANDO OS VALORES A E B ( Y = A + BX )

```python
model.fit(x_train, y_train)
```

```python
print('A (Intercept): ', model.intercept_)
print('B (Inclinação): ', model.coef_)
```

- Executando o modelo do dataset de teste

```python
predic = model.predict(x_test)
```

- Verificando valores reais de teste em relação aos valores preditos

```python
plt.scatter(y_test, predic);
```

O plot acima ainda não corresponde o que seria o ideal, pois quanto mais linear melhor o modelo é, mas seguimos.. 

```python
sns.displot(y_test - predic, bins = 32);
```

- Avaliando o modelo

```python
print('MAE (Erro Médio Absoluto): %.2f' %  metrics.mean_absolute_error(y_test, predic))
print('MSE (Erro Quadrático Médio): %.2f' %  metrics.mean_squared_error(y_test, predic))
print('RMSE (Raiz do Erro Quadrático Médio): %.2f' %  np.sqrt(metrics.mean_squared_error(y_test, predic)))
print('R2-Score: %.2f' % metrics.r2_score(predic, y_test))
```

Com o R² = 0.34, na verdade o modelo linear explica cerca de 34% da variância de HDI, Consumo de Cerveja e Vinho a partir do Happiness Score.

- Verificando os coeficientes

Podemos simplesmente verificar os coeficientes na intenção de gerar insights.

```python
coeff = pd.DataFrame(model.coef_, x.columns, columns = ['Coefs'])
coeff
```

# Conclusão: 
Ainda qua o modelo não seja um dos melhores, segundo a avaliação realizada, podemos inferir que:

- Mantendo as outras variáveis fixas, ao aumentar o HDI em 1 unidade, aumenta o Score de Felicidade em 0.006538. Pode ser concluído que o HDI do País colabora positivamente para o Happiness Score;


- Mantendo as outras variáveis fixas, ao aumentar o Beer_PerCapita em 1 unidade, aumenta o Score de Felicidade em 0.000286. Legal, o consumo de cerva deixa as pessoas alegres;


- Mantendo as outras variáveis fixas, ao aumentar o Wine_PerCapita em 1 unidade, o Score de Felicidade é reduzido em 0.001494 (ou aumentado em -0,001494). Ou seja, vinho entristece as pessoas.
